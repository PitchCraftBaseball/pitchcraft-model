```python
PITCH_FAMILIES = {
    'FF': 'fastball', 'SI': 'fastball', 'FC': 'fastball',
    'SL': 'breaking', 'CU': 'breaking', 'KC': 'breaking', 'SV': 'breaking', 'ST': 'breaking',
    'CH': 'offspeed', 'FS': 'offspeed', 'FO': 'offspeed',
}
FAMILY_KEYS = ['fastball', 'breaking', 'offspeed', 'other']

ABS_REMAPPING = {
    'automatic_ball': 'ABS',
    'automatic_strike': 'ABS',
}
```

```python
def build_season_arsenals(df, save_path=None):

    df = df[df["game_type"] == "R"].copy()
    df["pitch_type"] = (
        df["description"]
        .map(ABS_REMAPPING)
        .fillna(df["pitch_type"])
    )
    df = df[df["pitch_type"] != "ABS"]
    df = df[df["pitch_type"].notna()]

    all_arsenals = {}

    for (pitcher_id, season), group in df.groupby(["pitcher", "game_year"]):

        pitcher_key = str(int(pitcher_id))
        season_key  = str(int(season))

        num_pitches = len(group)

        pitch_arsenal = group["pitch_type"].value_counts()

        pitch_rates = {
            pitch: round(count / num_pitches, 4)
            for pitch, count in pitch_arsenal.items()
        }

        family_rates = {k: 0.0 for k in FAMILY_KEYS}
        for pitch, rate in pitch_rates.items():
            family = PITCH_FAMILIES.get(pitch, 'other')
            family_rates[family] += rate

        family_rates = {k: round(v, 4) for k, v in family_rates.items()}

        season_entry = {
            "total_pitches": num_pitches,
            "pitches": pitch_rates,
            "families": family_rates,
            "arsenal_mask": list(pitch_rates.keys()),
        }

        if pitcher_key not in all_arsenals:
            all_arsenals[pitcher_key] = {
                "player_name": group["player_name"].iloc[0]
            }

        all_arsenals[pitcher_key][season_key] = season_entry

        if save_path:
            with open(save_path, "w") as f:
                json.dump(all_arsenals, f, indent=4)

    return all_arsenals            
```

```python
df = pd.concat([data24, data25])
arsenals = build_season_arsenals(df, save_path="arsenals_all.json")

```
